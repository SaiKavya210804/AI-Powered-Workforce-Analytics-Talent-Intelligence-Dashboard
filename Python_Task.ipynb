```python
import pandas as pd

data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Amit", "Riya", "Rahul", "Sneha", "Kiran"],
    "Department": ["HR", "IT", "Finance", "IT", "Sales"],
    "Age": [25, 30, 28, 35, 27],
    "Salary": [40000, 55000, 50000, 70000, 45000]
}

df = pd.DataFrame(data)

print("Employee Data")
print(df)

print("\nAverage Age:", df["Age"].mean())

print("Average Salary:", df["Salary"].mean())

print("\nHighest Salary:")
print(df[df["Salary"] == df["Salary"].max()])

print("\nLowest Salary:")
print(df[df["Salary"] == df["Salary"].min()])

print("\nDepartment-wise Employee Count:")
print(df["Department"].value_counts())

print("\nTotal Employees:", len(df))

print("\nEmployees with Salary greater than 50000:")
print(df[df["Salary"] > 50000])

print("\nAverage Salary by Department:")
print(df.groupby("Department")["Salary"].mean())

print("\nEmployee Names:")
print(df["Name"])
```
