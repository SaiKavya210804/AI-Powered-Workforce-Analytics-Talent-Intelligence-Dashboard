```python
import sqlite3
import pandas as pd

# Connect to an in-memory SQLite database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# SQL DDL and DML statements
setup_sql = """
CREATE TABLE Employees (
    Employee_ID INT PRIMARY KEY,
    Name VARCHAR(50),
    Department VARCHAR(30),
    Age INT,
    Salary INT
);

INSERT INTO Employees VALUES
(101,'Amit','HR',25,40000),
(102,'Riya','IT',30,55000),
(103,'Rahul','Finance',28,50000),
(104,'Sneha','IT',35,70000),
(105,'Kiran','Sales',27,45000);
"""

# Execute setup SQL
cursor.executescript(setup_sql)
conn.commit()
print("Database and table 'Employees' created, and data inserted successfully.\n")

# List of SQL queries to execute and display results
queries = [
    ("Display All Employees", "SELECT * FROM Employees;"),
    ("Average Salary", "SELECT AVG(Salary) AS Average_Salary FROM Employees;"),
    ("Highest Salary", "SELECT MAX(Salary) AS Highest_Salary FROM Employees;"),
    ("Lowest Salary", "SELECT MIN(Salary) AS Lowest_Salary FROM Employees;"),
    ("Department-wise Employee Count", "SELECT Department, COUNT(*) AS Employee_Count FROM Employees GROUP BY Department;"),
    ("Employees with Salary Greater Than 50000", "SELECT * FROM Employees WHERE Salary > 50000;"),
    ("Employees Older Than 28", "SELECT * FROM Employees WHERE Age > 28;")
]

# Execute each query and print results
for title, query in queries:
    print(f"--- {title} ---")
    try:
        df = pd.read_sql_query(query, conn)
        print(df.to_markdown(index=False))
    except pd.io.sql.DatabaseError as e:
        print(f"Error executing query: {e}")
    print("\n")

# Close the connection
conn.close()
```
